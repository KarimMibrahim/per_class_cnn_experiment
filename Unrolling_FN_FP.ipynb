## For class sleep

```python
"""
This script is used to analyse the confusion in false postives and negatives
It looks up which playlists are most misclassified and which classes are confused together 
"""

import os 
import pandas as pd

#######################
# Read false negatives 
#######################
sleep_fn = pd.read_csv("experiments/C2_adadelta_30secs_Results/sleep_Per_class_/2019-04-23_15-01-40/sleep_false_negatives.txt", header = None)
sleep_fn.columns = ['track']
sleep_fn.track = [x.replace("https://www.deezer.com/en/track/","") for x in sleep_fn.track]

#######################
# Read ground truth
#######################
groundTruth = pd.read_pickle("GroundTruth/ground_truth_track_count_vector.pkl")
groundTruth = pd.read_pickle("GroundTruth/ground_truth_hot_vector.pkl")
groundTruth.drop("playlists_count",inplace = True,axis = 1 )

##############################################
# Check FN songs are sharing which other labels? 
##############################################
sleep_fn_groundtruth  = groundTruth[groundTruth.song_id.isin(sleep_fn.track)]
print("False negatives for class sleep share the following labels: ")
print(sleep_fn_groundtruth.sum().sort_values(ascending=False))

################################################
# Check FN songs come most from which playlists  
################################################
playlist_songs = pd.read_csv("GroundTruth/playlists_songs.tsv", sep  = "\t")
playlists_fn = playlist_songs[playlist_songs.song_id.isin(sleep_fn.track)]
playlist_tags = pd.read_csv("GroundTruth/sample_playlists_title.tsv", sep = "\t")
playlist_tags.drop("editor",inplace = True,axis = 1 )
playlist_fn_keywords = playlists_fn.merge(playlist_tags, on = 'playlist_id' , how = 'left')
print("\n\nPlaylists where most of the False negatives for class sleep come from are: ")
print(playlist_fn_keywords.groupby(["playlist_id",'matches', 'title']).count().sort_values(by = 'song_id',ascending=False).head())


"""
False positives
"""


#######################
# Read false postivies 
#######################
sleep_fp = pd.read_csv("experiments/C2_adadelta_30secs_Results/sleep_Per_class_/2019-04-23_15-01-40/sleep_false_positives.txt", header = None)
sleep_fp.columns = ['track']
sleep_fp.track = [x.replace("https://www.deezer.com/en/track/","") for x in sleep_fp.track]

#######################
# Read ground truth
#######################
groundTruth = pd.read_pickle("GroundTruth/ground_truth_track_count_vector.pkl")
groundTruth = pd.read_pickle("GroundTruth/ground_truth_hot_vector.pkl")
groundTruth.drop("playlists_count",inplace = True,axis = 1 )

##############################################
# Check fp songs are sharing which other labels? 
##############################################
sleep_fp_groundtruth  = groundTruth[groundTruth.song_id.isin(sleep_fp.track)]
print("\n\nFalse Positives for class sleep share the following labels: ")
print(sleep_fp_groundtruth.sum().sort_values(ascending=False))

################################################
# Check fp songs come most from which playlists  
################################################
playlist_songs = pd.read_csv("GroundTruth/playlists_songs.tsv", sep  = "\t")
playlists_fp = playlist_songs[playlist_songs.song_id.isin(sleep_fp.track)]
playlist_tags = pd.read_csv("GroundTruth/sample_playlists_title.tsv", sep = "\t")
playlist_tags.drop("editor",inplace = True,axis = 1 )
playlist_fp_keywords = playlists_fp.merge(playlist_tags, on = 'playlist_id' , how = 'left')
print("\n\nPlaylists where most of the False Positives for class sleep come from are: ")
print(playlist_fp_keywords.groupby(["playlist_id",'matches', 'title']).count().sort_values(by = 'song_id',ascending=False).head())



"""
True positives
"""


#######################
# Read True postivies 
#######################
sleep_tp = pd.read_csv("experiments/C2_adadelta_30secs_Results/sleep_Per_class_/2019-04-23_15-01-40/sleep_true_positives.txt", header = None)
sleep_tp.columns = ['track']
sleep_tp.track = [x.replace("https://www.deezer.com/en/track/","") for x in sleep_tp.track]

#################################################
# Check what trues positives share labels
#################################################

sleep_tp_groundtruth  = groundTruth[groundTruth.song_id.isin(sleep_tp.track)]
print("\n\nTrue Positives for class sleep share the following labels: ")
print(sleep_tp_groundtruth.sum().sort_values(ascending=False))
```
